```python
import pandas as pd
from src.pipeline.feature_engineering import FeaturePipeline
from src.pipeline.proxy_target import ProxyTargetEngineer
from src.pipeline.train_model import train_models
from src.pipeline.register_model import register_best_model
```

```python

def main():
    # Step 1: Load raw dataset
    raw_df = pd.read_csv("data/raw/transactions.csv")

    # Step 2: Apply feature engineering
    feature_pipeline = FeaturePipeline()
    X_processed = feature_pipeline.fit_transform(raw_df)
    X_processed.to_csv("data/processed/processed_features.csv", index=False)

    # Step 3: Generate proxy target variable
    proxy_engineer = ProxyTargetEngineer(snapshot_date="2023-12-31")
    target_df = proxy_engineer.generate_target(raw_df)

    # Step 4: Merge features with labels
    final_df = X_processed.merge(target_df, on="CustomerId", how="left")
    final_df.to_csv("data/processed/final_dataset.csv", index=False)

    # Step 5: Train models
    best_model, best_score, leaderboard = train_models(final_df)

    # Save leaderboard for comparison
    leaderboard.to_csv("data/processed/model_leaderboard.csv", index=False)

    # Step 6: Register best model
    register_best_model(best_model, score=best_score)


if __name__ == "__main__":
    main()
```
